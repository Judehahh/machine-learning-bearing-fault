# 短时傅里叶变换

```python
np.asarray(data,dtype = 'float')
```

```python
fs = 48000
N = 2000
amp = 2 * np.sqrt(2)

# 计算并绘制STFT的大小
f, t, Zxx = scipy.signal.stft(np.asarray(data,dtype = 'float')[:,0], fs, nperseg=1000)
plt.figure(figsize=(8, 4))
plt.tick_params(labelsize=15)
plt.pcolormesh([0,500,1000,1500,2000], f, np.abs(Zxx), vmin = 0, vmax = amp)
plt.ylabel(u"频率(Hz)", fontsize=15)
plt.xlabel(u"样本点数目", fontsize=15)
plt.show()
```

# 连续小波变换

```python
wavename = 'morl'
totalscal = 256
fc = pywt.central_frequency(wavename)
cparam = 2 * fc * totalscal
scales = cparam / np.arange(totalscal, 1, -1)
[cwtmatr, frequencies] = pywt.cwt(data, scales, wavename, 1.0/fs)
```

```python
plt.figure(figsize=(8, 4))
plt.xlim(0,2000)
plt.contourf(range(0, 2000), frequencies, abs(cwtmatr).reshape(255,2000))
plt.tick_params(labelsize=15)
plt.ylabel(u"频率(Hz)", fontsize=15)
plt.xlabel(u"样本点数目", fontsize=15)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

# EMD分解

```python
#EMD处理
emd = EMD()
emd.emd(np.asarray(data,dtype = 'float')[:,0])
imfs, res = emd.get_imfs_and_residue()
# 绘制 IMF
vis = Visualisation()
```

```python
n = 1
plt.figure(figsize=(20,15))
plt.subplots_adjust(wspace = 0, hspace = 0.5)
for imf in imfs:
    plt.subplot(len(imfs[:,0]),1,n)
    plt.xlim(0,2000)
    plt.tick_params(labelsize=15)
    plt.plot(range(2000), imf)
    plt.title("IMF "+str(n),fontsize=20)
    plt.xlabel("")
    n += 1
```
